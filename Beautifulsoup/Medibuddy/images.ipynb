```python
import cv2
import numpy as np
import random
import os
import glob

def random_color_adjustment(image):
    """Apply random color adjustment to the image."""
    # Randomly adjust brightness, contrast, and saturation
    alpha = random.uniform(0.7, 1.3)  # Brightness/Contrast adjustment factor
    beta = random.randint(-20, 20)    # Brightness adjustment value
    saturation_factor = random.uniform(0.7, 1.3)  # Saturation adjustment factor

    # Convert image to HSV color space
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Apply adjustments
    hsv_image[:, :, 2] = np.clip(alpha * hsv_image[:, :, 2] + beta, 0, 255)
    hsv_image[:, :, 1] = np.clip(saturation_factor * hsv_image[:, :, 1], 0, 255)

    # Convert back to BGR color space
    adjusted_image = cv2.cvtColor(hsv_image, cv2.COLOR_HSV2BGR)
    return adjusted_image

def make_image_darker(image):
    """Reduce the brightness of the image to make it darker."""
    brightness_factor = random.uniform(0.8, 1)  # Adjust this for darkness level
    darkened_image = (image * brightness_factor).astype(np.uint8)
    return darkened_image


def add_watermark(input_image, watermark_image_path, scale_factor = 0.5):
    # Read the input image and watermark image
    
    watermark = cv2.imread(watermark_image_path, cv2.IMREAD_UNCHANGED)

    # Get the dimensions of the watermark image
    wm_h, wm_w = watermark.shape[:2]

    # Resize the watermark based on the scale factor
    watermark = cv2.resize(watermark, (int(wm_w * scale_factor), int(wm_h * scale_factor)))

    # Get the updated dimensions of the watermark
    wm_h, wm_w = watermark.shape[:2]
    
    # Define the position to place the watermark (top-left corner)
    position = (10, 10)  # Adjust the position as needed

    # Extract the alpha channel (transparency) from the watermark
    alpha_channel = watermark[:, :, 3]

    # Create a region of interest (ROI) in the input image for placing the watermark
    roi = input_image[position[1]:position[1] + wm_h, position[0]:position[0] + wm_w]
    
    print (roi)
    print (wm_h, wm_w)
    
    # Blend the resized watermark with the ROI using the alpha channel
    for c in range(0, 3):
        print (watermark[:, :, c])
        input_image[position[1]:position[1] + wm_h, position[0]:position[0] + wm_w, c] = \
            input_image[position[1]:position[1] + wm_h, position[0]:position[0] + wm_w, c] * \
            (1 - alpha_channel / 255.0) + watermark[:, :, c] * (alpha_channel / 255.0)

    return input_image


# Folder path containing image files
folder_path = "/Users/vinaykumar/Downloads/medibuddy_images"

# List all image files in the folder (supports common image formats like JPG, PNG, etc.)
image_files = glob.glob(os.path.join(folder_path, "*.webp"))

# Path to the watermark image
watermark_path = "/Users/vinaykumar/Downloads/medibuddy_logo.png"

# Process each image
for image_path in image_files:
    img = cv2.imread(image_path)

    # Apply random color adjustment
    img = random_color_adjustment(img)

    # Make the image darker
    img = make_image_darker(img)

    # Add watermark to the image
    img_with_watermark = add_watermark(img, watermark_path)

    # Save the modified image with watermark in the same folder
    filename = os.path.basename(image_path)
    output_image_path = os.path.join(folder_path, filename.replace(".webp", "_modified_with_watermark.jpg"))
    cv2.imwrite(output_image_path, img_with_watermark)
```
